```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix

# Đường dẫn tới tệp dữ liệu
FILE_PATH = "/content/drive/My Drive/Machine Learning/Dataset/Admission_Predict.csv"

# Đọc dữ liệu
data = pd.read_csv(FILE_PATH)

# Đổi tên cột để dễ xử lý hơn
data.rename(columns={"Chance of Admit ": "Chance_of_Admit"}, inplace=True)

# Gắn nhãn nhị phân: 1 nếu >= threshold, -1 nếu < threshold
THRESHOLD = 0.7
data['Label'] = np.where(data['Chance_of_Admit'] >= THRESHOLD, 1, -1)

# Chuẩn bị dữ liệu
X = data.drop(columns=['Serial No.', 'Chance_of_Admit', 'Label']).values  # Đặc trưng
y = data['Label'].values  # Nhãn

# Chia tập dữ liệu thành huấn luyện và kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Tạo và huấn luyện mô hình SVM
C = 1.0
svm_model = SVC(kernel='linear', C=C)
svm_model.fit(X_train, y_train)
```

```python
# Đánh giá mô hình
def evaluate_model(model, X, y, dataset_name="Dataset"):
    predictions = model.predict(X)
    accuracy = accuracy_score(y, predictions)
    conf_matrix = confusion_matrix(y, predictions)
    print(f"\nĐánh giá trên {dataset_name}:")
    print(f"Accuracy: {accuracy * 100:.2f}%")
    print("Confusion Matrix:")
    print(conf_matrix)
    return accuracy, conf_matrix

# Đánh giá trên tập huấn luyện
evaluate_model(svm_model, X_train, y_train, "tập huấn luyện")

# Đánh giá trên tập kiểm tra
evaluate_model(svm_model, X_test, y_test, "tập kiểm tra")
```
