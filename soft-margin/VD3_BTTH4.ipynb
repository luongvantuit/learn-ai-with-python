```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Load dataset
data_path = r"/content/drive/My Drive/Machine Learning/Dataset/spambase.data"

# Define column headers (57 features + 1 label column)
columns = [f"feature_{i}" for i in range(1, 58)] + ["label"]
data = pd.read_csv(data_path, header=None, names=columns)

# Map labels: 1 -> 1 (Spam), 0 -> -1 (Non-spam)
data["label"] = data["label"].map({1: 1, 0: -1})

# Split features and labels
X = data.iloc[:, :-1].values  # All columns except the label column
y = data.iloc[:, -1].values   # Label column

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
# Initialize and train the SVM model
C = 1.0
model = SVC(kernel="linear", C=C)
model.fit(X_train, y_train)
```

```python
# Evaluate on training set
y_train_pred = model.predict(X_train)
accuracy_train = accuracy_score(y_train, y_train_pred)
conf_matrix_train = confusion_matrix(y_train, y_train_pred)

# Evaluate on testing set
y_test_pred = model.predict(X_test)
accuracy_test = accuracy_score(y_test, y_test_pred)
conf_matrix_test = confusion_matrix(y_test, y_test_pred)

# Print results
print("\nEvaluation on Training Set:")
print(f"Accuracy: {accuracy_train * 100:.2f}%")
print("Confusion Matrix:")
print(conf_matrix_train)

print("\nEvaluation on Testing Set:")
print(f"Accuracy: {accuracy_test * 100:.2f}%")
print("Confusion Matrix:")
print(conf_matrix_test)
```
